## Train | Test Split and Scaling

```python
X = df.drop('test_result',axis=1)
y = df['test_result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
# scaling
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) # scale fit işlemi sadece X_train' e, transform ise X_train ve X_test'e uygulanır
X_test_scaled = scaler.transform(X_test)
```

## Modelling

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()                       # model tanımlandı
```

```python
log_model.fit(X_train_scaled, y_train)                 # fit işlemi 
```

```python
log_model.coef_                                        # katsayılar

# katsayıların durumuna göre physical_score sonucu daha fazla etkiliyor
```

```python
log_model.intercept_                                   # sabit
```

```python
df.head()
```

```python
y_pred = log_model.predict(X_test_scaled)   # X_test_scaled ile tahmin yapıyoruz
y_pred
```

```python
y_pred_proba = log_model.predict_proba(X_test_scaled)
y_pred_proba

# modele göre yapılan tahmin de herbir observation için 0 ve 1 olma olasılığını gösterir. 
# sol taraf 0 olma olasılığı sağ taraf 1 olma olasılığı. olasılıklar toplamı 1 dir
```

```python
test_data = pd.concat([X_test, y_test], axis=1)    # bugun görmek için yapıyoruz sonrasında yapmıyoruz
test_data
```

```python
test_data["pred_proba"] = y_pred_proba[:,1]
test_data
```

```python
test_data["pred"] = y_pred
test_data
```

son 3 kod satırı ile X_test, y_test, test_result y_pred_proba(1 olma olasılığı) ve  y_pred durumlarını birlikte görmek istedik. model ne kadar başarılı tahminler yapmış incelemek istedik.

2371 index numaralı observation a bakarsak test_result yani gerçek değerimiz 0 olmasına rağmen yüksek olasılık değeri ile 1 tahmininde bulunmuş. başarılı tahminlerde gayet net bir şekilde görülüyor

```python
print(classification_report(y_test, y_pred))
```

* support : 0 ve 1 lerin sayısını verir
* accuracy :  doğru tahminlerin (1 ve 0) tamamının tahmin yapılan dataya oranı. unbalance durumunda bakılmaz
* recall (sensitivity) : 1 olarak tahmin edilen ve gerçek değerleri 1 olanların yani TruePozitiflerin, gerçek değeri 1 olanlara yani FalseNegatif ve TruePozitiflerin toplamına oranı (gerçek değerlerin kaçtakaçını bildim?)
* precision : 1 olarak tahmin edilen ve gerçek değerleri 1 olanların yani TruePozitiflerin, tahmin edilen değeri 1 olanlara yani FalsePozitif ve TruePozitiflerin toplamına oranı (tahminlerimin kaçtakaçı doğru?)
* f1 score = precision ve recall un harmonik ortalamasıdır. unbalance durumda hatalı sonuç vermesini engeller

* recall ve f1 score un artmasına yönelik çabamız olacak

örnek
* precision = 0.1, recall = 1 ve kanser olan hasta sayısı 100 ise nasıl yorum yapılır?

recall değerinin 1 olması gerçekte hasta olanların tamamının bilindiğini gösterir. precision değerinin 0.1 olması ise 100 hastayı bilmek için 1000 tahminde bulunduğumuzu gösterir. bu nedenle modelin başarısız olduğu söylenebilir.
score değerlerinin birbirine yakın olması iyidir
* precision düşük recall yüksekse model sallıyor demektir. tersi durumda ise model düzgün tahmin yapmıyor demektir. ikisi arasındaki uyum önemlidir

sensetive = recall
