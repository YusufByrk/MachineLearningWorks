## Train | Test Split and Scaling

```python
X = df.drop('test_result',axis=1)
y = df['test_result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
# scaling
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train) # scale fit ilemi sadece X_train' e, transform ise X_train ve X_test'e uygulanr
X_test_scaled = scaler.transform(X_test)
```

## Modelling

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()                       # model tanmland
```

```python
log_model.fit(X_train_scaled, y_train)                 # fit ilemi 
```

```python
log_model.coef_                                        # katsaylar

# katsaylarn durumuna gre physical_score sonucu daha fazla etkiliyor
```

```python
log_model.intercept_                                   # sabit
```

```python
df.head()
```

```python
y_pred = log_model.predict(X_test_scaled)   # X_test_scaled ile tahmin yapyoruz
y_pred
```

```python
y_pred_proba = log_model.predict_proba(X_test_scaled)
y_pred_proba

# modele gre yaplan tahmin de herbir observation iin 0 ve 1 olma olasln gsterir. 
# sol taraf 0 olma olasl sa taraf 1 olma olasl. olaslklar toplam 1 dir
```

```python
test_data = pd.concat([X_test, y_test], axis=1)    # bugun grmek iin yapyoruz sonrasnda yapmyoruz
test_data
```

```python
test_data["pred_proba"] = y_pred_proba[:,1]
test_data
```

```python
test_data["pred"] = y_pred
test_data
```

son 3 kod satr ile X_test, y_test, test_result y_pred_proba(1 olma olasl) ve  y_pred durumlarn birlikte grmek istedik. model ne kadar baarl tahminler yapm incelemek istedik.

2371 index numaral observation a bakarsak test_result yani gerek deerimiz 0 olmasna ramen yksek olaslk deeri ile 1 tahmininde bulunmu. baarl tahminlerde gayet net bir ekilde grlyor

```python
print(classification_report(y_test, y_pred))
```

* support : 0 ve 1 lerin saysn verir
* accuracy :  doru tahminlerin (1 ve 0) tamamnn tahmin yaplan dataya oran. unbalance durumunda baklmaz
* recall (sensitivity) : 1 olarak tahmin edilen ve gerek deerleri 1 olanlarn yani TruePozitiflerin, gerek deeri 1 olanlara yani FalseNegatif ve TruePozitiflerin toplamna oran (gerek deerlerin katakan bildim?)
* precision : 1 olarak tahmin edilen ve gerek deerleri 1 olanlarn yani TruePozitiflerin, tahmin edilen deeri 1 olanlara yani FalsePozitif ve TruePozitiflerin toplamna oran (tahminlerimin kataka doru?)
* f1 score = precision ve recall un harmonik ortalamasdr. unbalance durumda hatal sonu vermesini engeller

* recall ve f1 score un artmasna ynelik abamz olacak

rnek
* precision = 0.1, recall = 1 ve kanser olan hasta says 100 ise nasl yorum yaplr?

recall deerinin 1 olmas gerekte hasta olanlarn tamamnn bilindiini gsterir. precision deerinin 0.1 olmas ise 100 hastay bilmek iin 1000 tahminde bulunduumuzu gsterir. bu nedenle modelin baarsz olduu sylenebilir.
score deerlerinin birbirine yakn olmas iyidir
* precision dk recall yksekse model sallyor demektir. tersi durumda ise model dzgn tahmin yapmyor demektir. ikisi arasndaki uyum nemlidir

sensetive = recall
