## Spliting the dataset into X (independent variable) and y (dependent variable)

Independent variables = X

Dependent variables   = y

y = b1 * X + b0

y = baml deiken = target variable = dependent variable

b1 = katsay/eim, slope 0 < b1 < 1 and -1 < b1 < 0

X = bamsz deiken /feature, independent variable

b0 = intercept /otonom parametre / balang parametresi - < 0 < +

```python
X = df["total_spend"]
y = df["sales"]
```

```python
np.polyfit(X, y, deg =1) # simple linear reg olduu iin bu kodu kullandk ama normalde hep skitlearn kullanacaz.
```

* deg = 1, polinomun derecesini (yani 1. derece bir polinom oldugunu) gsteriyor. Kisaca lineer fit.
* b1 = 0.048  b0 = 4.24 verir

## Determination of coofficients (slopes, intercept)

```python
slope, intercept = np.polyfit(X, y, deg =1)     # burada X' e slope, y' ye intercept atanyor
```

```python
print("slope     :", slope) # feature nin katsay deeri
print("intercept :", intercept) # balang parametresi yani herhangi bir reklam verilmese de sales 4.243 eklinde 
```

```python
b1, b0 = np.polyfit(X, y, deg=1)
```

```python
y_pred = b1 * X + b0
```

```python
y_pred
```

```python
values = {"actual": y, "predicted":y_pred, "resudial": y - y_pred, "LSE" : (y - y_pred)**2}
df_2 = pd.DataFrame(values)
df_2
```

* actual : gerek y deerleri
* predicted : tahmin y deerleri
* resudial : (y - y_pred) hata
* LSE : en kk karaler hatas

```python
df_2.resudial.sum().round()
```

```python
df_2.LSE.sum()
```

noktalarin cizgiye olan uzakliklarinin kareleri toplamini cikan sonuc 1338 den daha kucuk yapmak mumkun degil. Olabilecek minumum deger 1338

## Prediction with simple linear regression

* 0-500 aras 100 tane reklam miktar retip modelin bu deerler iin tahmin edilen sat deerleri nelerdir
* prediction modelimizi deniyoruz burada. ncelikle X deerleri retiyoruz potantial_spend olarak.
* sonrasndada y_tahmin deeri olarak predicted_sales_lin buluyoruz

```python
potential_spend = np.linspace(0, 500, 100) # potansiyel harcama
potential_spend
```

```python
predicted_sales_lin = b1 * potential_spend + b0
predicted_sales_lin
```
