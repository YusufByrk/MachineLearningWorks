## Train test split

```python
X=df.drop("Selling_Price", axis=1)
y=df.Selling_Price
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=5)

print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

## Modeling

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
DT_model = DecisionTreeRegressor(random_state=101)
```

```python
DT_model.fit(X_train, y_train)
```

```python
y_pred = DT_model.predict(X_test)
y_train_pred = DT_model.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"train": {"R2" : r2_score(y_train, y_train_pred),
                        "mae" : mean_absolute_error(y_train, y_train_pred),
                        "mse" : mean_squared_error(y_train, y_train_pred),                          
                        "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
              "test": {"R2" : r2_score(y_test, y_pred),
                       "mae" : mean_absolute_error(y_test, y_pred),
                       "mse" : mean_squared_error(y_test, y_pred),
                       "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
train_val(DT_model, X_train, y_train, X_test, y_test)
```

train setimizde hata yok. score lar optimum score lar. test setimizde isi sonuçlar bu şekilde görünmüyor. overfitting olduğunu anlıyoruz

```python
# CV ile score lara tekrar bakalım

from sklearn.model_selection import cross_validate, cross_val_score

model = DecisionTreeRegressor(random_state=101)

scores = cross_validate(model, X_train, y_train, scoring=['r2', 
                                                          'neg_mean_absolute_error',
                                                          'neg_mean_squared_error',
                                                          'neg_root_mean_squared_error'], cv =5)

df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

overfit olduğunu anlıyoruz

## Visualizing trees

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8), dpi=150)
plot_tree(DT_model, filled=True, feature_names=X.columns);
plt.show()
```

```python
def report_model(model):                     # classification model olmadığı için score lar değişti, R2, mae, mse, rmse
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    print('\n')
    
    print(train_val(model, X_train, y_train, X_test, y_test))
    print('\n')
    
    plt.figure(figsize=(12,8),dpi=100)
    plot_tree(model,filled=True, feature_names=X.columns, fontsize=10);
```

* max_depth 3 yaparak budama (pruned) yaparak score lara bakıyoruz

```python
pruned_tree = DecisionTreeRegressor(max_depth=3, random_state=101)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
plt.show()
```

* kök ve node lardaki ayrıştırma varyans üzerinden yapılıyor.

## GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = DecisionTreeRegressor(random_state=101)
```

```python
param_grid = {"splitter":["best", "random"],
              "max_depth": [None, 4, 5, 6, 7,8,9,10],
              "min_samples_leaf": [1, 2, 3, 4, 5, 6,7],
              "min_samples_split": [2, 3, 5, 6, 7,8,9]}

# gridsearch ile teyit ediyoruz DT regression ile rmse değerleri uyumlu
```

```python
grid_model = GridSearchCV(estimator=model,
                          param_grid=param_grid,
                          scoring='neg_root_mean_squared_error',
                          cv=10,
                          n_jobs = -1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_                # best parametreleri bulduk.
```

```python
grid_score =pd.DataFrame(grid_model.cv_results_)
grid_score
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_         # neg_root_mean_squared_error
```

#### grid_model ile y_train_pred buluyoruz

```python
y_train_pred = grid_model.predict(X_train)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
train_val(grid_model, X_train, y_train, X_test, y_test)
```

* ilk score larımızla karşılaştırdığımızda grid_model imizin score ları başarılı
