## Train test split

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
X=df.drop("Selling_Price", axis=1)
y=df.Selling_Price
```

__Encoder__

Modelimiz tree based bir model oldugu icin dummy yerine encoder islemini uygulayacagiz :

```python
cat = X.select_dtypes("object").columns
cat
```

```python
X[cat].head()
```

```python
enc = OrdinalEncoder()
X[cat] = enc.fit_transform(X[cat])
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=5)
```

```python
print("Train features shape : ", X_train.shape)
print("Train target shape   : ", y_train.shape)
print("Test features shape  : ", X_test.shape)
print("Test target shape    : ", y_test.shape)
```

## Modeling for Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
DT_model = DecisionTreeRegressor(random_state=101)
```

```python
DT_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def train_val(model, X_train, y_train, X_test, y_test):
    
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    
    scores = {"train": {"R2" : r2_score(y_train, y_train_pred),
                        "mae" : mean_absolute_error(y_train, y_train_pred),
                        "mse" : mean_squared_error(y_train, y_train_pred),                          
                        "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
              "test": {"R2" : r2_score(y_test, y_pred),
                       "mae" : mean_absolute_error(y_test, y_pred),
                       "mse" : mean_squared_error(y_test, y_pred),
                       "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
train_val(DT_model, X_train, y_train, X_test, y_test)
```

Regression modellerinde train datasının bu şekilde sonuç vermesi mümkün değil. Overfit olduğunu gösterir.
R2 nin 1 diğer score ların 0 olduğu duruma pure deniyor ve mümkün değil. tüm score lar max değer almış olması purity saflık.

__Cross Validate__

```python
from sklearn.model_selection import cross_validate, cross_val_score
model = DecisionTreeRegressor(random_state=101)
scores = cross_validate(model, X_train, y_train, scoring=['r2', 
            'neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error'], cv =10)
df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

## Visualizing trees

```python
from sklearn.tree import plot_tree

def report_model(model):
    y_pred = model.predict(X_test)
    y_train_pred = model.predict(X_train)
    print('\n')
    print(train_val(model, X_train, y_train, X_test, y_test))
    print('\n')
    plt.figure(figsize=(12,8),dpi=100)
    plot_tree(model.estimators_[0],filled=True, feature_names=X.columns, fontsize=10);
```

```python
pruned_tree = RandomForestRegressor(n_estimators=250, max_depth=4, random_state=101)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

burada dikkat edilmesi gerekn bir durum var. Max_fetaure parametresinin çalışma yöntemi burada default değeri olarak aldığı auto değeri RFClassifier dan farklı çalışıyor. yani RFClassifier ile RFRegressior default değerleri farklı çalışıyor
* RFClassifier da default değer auto olduğu zaman fetaure sayısının karekökünü alarak random feature seçiyor.
* RFRegressior de ise default değer olan 1 ile toplam feature sayısı baz alınarak random feature seçiyor. bu yüzden en değerli feature kök ve ilk node larda aynı. shift+tab ile bak

## GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={ 'n_estimators': [64, 128, 200, 500]
             'max_depth': [None, 4, 5, 6, 7, 8],
             'max_features': ['auto', len(df.columns)/3, 2, 4],    # RFReggressior de max_features için len(df.columns)/3 kullanmak
             'min_samples_leaf': [1, 2, 3, 4],                     # faydalı tavsiye ediliyor. feature sayısının 1/3 ü
             'min_samples_split': [2, 3, 5, 6]}

model = RandomForestRegressor(random_state=101)

grid_model = GridSearchCV(estimator=model,
                          param_grid = param_grid,
                          scoring='neg_root_mean_squared_error'
                          cv=10,
                          n_jobs=-1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
grid_score =pd.DataFrame(grid_model.cv_results_)
grid_score
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_                  # GridSearc için score olarak rmse seçtiğimiz için rmse score dur
```

```python
train_val(grid_model, X_train, y_train, X_test, y_test)
```

__Cross Validate__

GridSearch sonrası CV gereksizdir. GS zaten kendi içinde CV yapıyor. score lar aynı çıkar görmek için yapıyoruz. rmse değeri GridSearch best score ile karşılaştırınca göreceğiz

```python
from sklearn.model_selection import cross_validate, cross_val_score

model = RandomForestRegressor(random_state=101, max_depth = None,
             max_features= 4,
             min_samples_leaf = 1,
             min_samples_split= 2,
             n_estimators= 200)
scores = cross_validate(model, X_train, y_train, scoring=['r2', 
            'neg_mean_absolute_error','neg_mean_squared_error','neg_root_mean_squared_error'], cv = 10)
df_scores = pd.DataFrame(scores)
df_scores.mean()[2:]
```

#### Grid Search modelimiz ile bulduğumuz prediction değerler ile residualları görelim

```python
y_pred = grid_model.predict(X_test)
my_dict = { 'Actual': y_test, 'Pred': y_pred, 'Residual': y_test-y_pred }
compare = pd.DataFrame(my_dict)
```
