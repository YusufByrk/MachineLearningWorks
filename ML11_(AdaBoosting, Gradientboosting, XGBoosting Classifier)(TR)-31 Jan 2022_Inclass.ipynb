## Modelling and Model Performance

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

Bir çocuk on köpek görürse ve hepsinin rengi siyahsa, tüm köpeklerin siyah olduğuna inanabilir ve bir köpeği tanımlamaya çalışırken bunu bir özellik olarak düşünebilir.Ona beyaz bir köpek gösterildiğini ve ailesinin ona bunun bir köpek olduğunu söylediğini hayal edin. İstenen bir öğrenme oranıyla, siyah rengin köpeklerin önemli bir özelliği olmadığını çabucak anlayacak ve başka bir özellik arayacaktır.
Ancak düşük bir öğrenme oranıyla, beyaz köpeği bir aykırı değer olarak görür ve tüm köpeklerin siyah olduğuna inanmaya devam ederdi.
Ve eğer öğrenme oranı çok yüksekse, beyaz olanlardan daha fazla siyah köpek görmüş olmasına rağmen, anında tüm köpeklerin beyaz olduğuna inanmaya başlayacaktı.
Mesele şu ki, arzu edilen bir öğrenme oranına ulaşmak gerçekten önemli çünkü: hem düşük hem de yüksek öğrenme oranları zaman ve kaynak israfına neden olur.
Daha düşük bir öğrenme oranı, daha fazla eğitim süresi anlamına gelir daha fazla zaman, artan bulut GPU maliyetleriyle sonuçlanır daha yüksek bir oran, hiçbir şeyi doğru bir şekilde tahmin edemeyebilecek bir modelle sonuçlanabilir. İstenen bir öğrenme oranı, ağın yararlı bir şeye yakınsaması için yeterince düşük, ancak makul bir sürede eğitilebilmesi için yeterince yüksek olandır.

## Analyzing performance as more weak learners are added.

```python
error_rates = [] # data büyükse bu kodu çalıştarma, hoca sadece göstermek için bu data da küçük olduğu için koymuş

for n in range(1,100):
    
    model = AdaBoostClassifier(n_estimators=n)
    model.fit(X_train,y_train)
    preds = model.predict(X_test)
    err = 1 - f1_score(y_test,preds)
    
    error_rates.append(err)
```

```python
plt.plot(range(1,100), error_rates);
```

## Gridsearch

```python
from sklearn.model_selection import GridSearchCV

model = AdaBoostClassifier(random_state=42)
param_grid = {"n_estimators": [20, 30, 100, 200], "learning_rate": [0.01, 0.1, 0.2, 0.5, 1.0]}
ada_grid_model = GridSearchCV(model, param_grid, cv=5, scoring= 'f1')
ada_grid_model.fit(X_train, y_train)
```

* ``n_estimators:`` Bu parametre, AdaBoost algoritmasındaki toplam tahmin edici sayısını belirler. Yani, AdaBoost'un kaç adet zayıf tahmin edici kullanacağını belirler. Burada kullanılan değerler 20, 30, 100 ve 200'dür.

* ``learning_rate:`` Bu parametre, her bir tahmin edicinin katkısının nasıl düzenleneceğini belirler. Büyük bir öğrenme oranı, her bir tahmin edicinin katkısını daha büyük bir şekilde arttıracaktır. Burada kullanılan değerler 0.01, 0.1, 0.2, 0.5 ve 1.0'dır.


    - Bu parametrelerin kombinasyonları, GridSearchCV tarafından AdaBoost algoritmasının performansını ölçmek için kullanılacaktır. GridSearchCV aynı zamanda 5 katlı çapraz doğrulama (5-fold cross-validation) yapacak ve en iyi performansı veren parametreleri seçecektir. Son olarak, scoring parametresi "f1" olarak belirlenmiştir, bu da sınıflandırma performansının F1 puanı (F1 score) kullanılarak ölçüleceğini belirtir.

```python
ada_grid_model.best_params_
```

* leraning rate in küçük olması n_estimator büyük olması model başarısı için daha iyi

## Feature_importances

* Not: random forest ve xgboost haricinde feature importance doğru bi sonuç vermiyor.

* Feature importance scores can provide insight into the model. Most importance scores are calculated by a predictive model that has been fit on the dataset. Inspecting the importance score provides insight into that specific model and which features are the most important and least important to the model when making a prediction. This is a type of model interpretation that can be performed for those models that support it.
Feature importance can be used to improve a predictive model. This can be achieved by using the importance scores to select those features to delete (lowest scores) or those features to keep (highest scores). This is a type of feature selection and can simplify the problem that is being modeled, speed up the modeling process (deleting features is called dimensionality reduction), and in some cases, improve the performance of the model.

```python
model = AdaBoostClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
model.fit(X_train, y_train)
model.feature_importances_

feats = pd.DataFrame(index=X.columns, data= model.feature_importances_, columns=['ada_importance'])
ada_imp_feats = feats.sort_values("ada_importance")
ada_imp_feats
```

## Comparing Models

```python
compare = pd.DataFrame({"Model": ["Logistic Regression", "KNN", "SVM", "Decision Tree", "Random Forest", "AdaBoost",
                                 "GradientBoost", "XGBoost"],
                        "F1": [log_f1, knn_f1, svc_f1, dt_f1, rf_f1, ada_f1, gb_f1, xgb_f1],
                        "Recall": [log_recall, knn_recall, svc_recall, dt_recall, rf_recall, ada_recall, gb_recall, xgb_recall],
                        "ROC_AUC": [log_auc, knn_auc, svc_auc, dt_auc, rf_auc, ada_auc, gb_auc, xgb_auc]})

def labels(ax):
    for p in ax.patches:
        width = p.get_width()                        # get bar length
        ax.text(width,                               # set the text at 1 unit right of the bar
                p.get_y() + p.get_height() / 2,      # get Y coordinate + X coordinate / 2
                '{:1.3f}'.format(width),             # set variable to display, 2 decimals
                ha = 'left',                         # horizontal alignment
                va = 'center')                       # vertical alignment
    
plt.figure(figsize=(14,10))
plt.subplot(311)
compare = compare.sort_values(by="F1", ascending=False)
ax=sns.barplot(x="F1", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(312)
compare = compare.sort_values(by="Recall", ascending=False)
ax=sns.barplot(x="Recall", y="Model", data=compare, palette="Blues_d")
labels(ax)

plt.subplot(313)
compare = compare.sort_values(by="ROC_AUC", ascending=False)
ax=sns.barplot(x="ROC_AUC", y="Model", data=compare, palette="Blues_d")
labels(ax)
plt.show()
```
